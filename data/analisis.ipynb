```python
# Librerias necesarias
import pandas as pd
import json
```

```python
# Modificacion de links de restaurantes para agregar parametros necesarios y botar duplicados
actual = "https.www.ubereats.com"
replace = "https://www.ubereats.com"
restaurantes = pd.read_csv("restaurantes.csv")
restaurantes["service_link"] = restaurantes["service_link"] + "?diningMode=DELIVERY&surfaceName="
restaurantes["service_link"] = restaurantes["service_link"].str.replace(actual, replace)
urls = restaurantes["service_link"].drop_duplicates().reset_index(drop=True)
urls.to_csv("service_links.csv", index=False, header=True)
```
